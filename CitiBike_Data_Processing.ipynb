```python
from glob import glob
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
def countRiders(month, dataframe):
    d = {}
    df = dataframe.loc[dataframe['month'] == str(month)]
    d['Rider Total'] = len(df)
    d['Female Total'] = len(df.loc[df['gender'] == '2'])
    d['Male Total'] = len(df.loc[df['gender'] == '1'])
    d['Undisclosed Total'] = len(df.loc[df['gender'] == '0'])
    d['Annual Members'] = len(df.loc[df['usertype'] == 'Subscriber'])
    d['Non Members'] = len(df.loc[df['usertype'] == 'Customer'])

    return d
```

```python
def ridersPerMonth(dataframe):
    i = 1
    x =1
    months = []
    
    while i < 13:
        data = countRiders(x, dataframe)
        months.append(data)
        i += 1
        x += 1
        
    return months
```

```python
def bikeDataToDF(year):
    combined_csv = sorted(glob(f'data/{year}/*-citibike-tripdata.csv'))
    
    cbd_df = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in combined_csv), ignore_index = True)
    
    cbd_df['starttime'] = pd.to_datetime(cbd_df['starttime'])
    cbd_df['stoptime'] = pd.to_datetime(cbd_df['stoptime'])
    cbd_df['year'] = cbd_df['starttime'].dt.year
    cbd_df['month'] = cbd_df['starttime'].dt.month
    cbd_df['day'] = cbd_df['starttime'].dt.day

    cbd_df['month'] = cbd_df['month'].astype(str)
    cbd_df['day'] = cbd_df['day'].astype(str)
    cbd_df['gender'] = cbd_df['gender'].astype(str)
    
    return cbd_df
```

```python
def bikeDataToDataFrame(year):
    combined_csv = sorted(glob(f'data/{year}/*-citibike-tripdata.csv'))
    
    cbd_df = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in combined_csv), ignore_index = True)
    
    cbd_df['starttime'] = pd.to_datetime(cbd_df['starttime'])
    cbd_df['stoptime'] = pd.to_datetime(cbd_df['stoptime'])
#     cbd_df['year'] = cbd_df['starttime'].dt.year
#     cbd_df['month'] = cbd_df['starttime'].dt.month
#     cbd_df['day'] = cbd_df['starttime'].dt.day

#     cbd_df['month'] = cbd_df['month'].astype(str)
#     cbd_df['day'] = cbd_df['day'].astype(str)
#     cbd_df['gender'] = cbd_df['gender'].astype(str)
    
    return cbd_df
```

```python
def getYearlyRiderCount(year):
    df = pd.DataFrame(ridersPerMonth(bikeDataToDF(f'{year}')))
    df.insert(0, 'Year', f'{year}')
    df.insert(1, 'Month', range(1, 1 + len(df)))
    
    return df
```

```python
data_2020 = bikeDataToDataFrame(2020)
data_2020['bike_usage'] = data_2020.groupby('bikeid')['bikeid'].transform('count')
data_2020['start_station_count'] = data_2020.groupby('start station id')['start station id'].transform('count')
data_2020['end_station_count'] = data_2020.groupby('end station id')['end station id'].transform('count')
data_2020.head()
```

```python
start = data_2020.groupby('start station id')['start station id'].transform('count')
stop = data_2020.groupby('end station id')['end station id'].transform('count')
data_2020['total_usage'] = start + stop
data_2020.head()
```

```python
data_2020['total_trip_duration_(mins)'] = data_2020.groupby('bikeid')['tripduration'].transform('sum')
data_2020['total_trip_duration_(mins)'] = data_2020['total_trip_duration_(mins)'].div(60).round(2) 
data_2020.head()
```
